## Statistical Calculations

To evaluate the behavior of the structural and dynamic properties of the simulated molecules, several statistical metrics were calculated, providing insight into the stability and fluctuations of the properties over time. The properties studied were:

- **RMSD** (Root Mean Square Deviation)
- **RMSF** (Root Mean Square Fluctuation)
- **Radius of Gyration** (Rg)
- **HBond** (Number of Hydrogen Bonds)
- **SASA** (Solvent Accessible Surface Area)

For each property, three key values were calculated:

1. **Mean (Average)**:  
   This value represents the central or typical behavior of each property throughout the simulation. It is calculated with the formula:  
   $$ \text{Mean} = \frac{1}{n} \sum_{i=1}^{n} x_i $$  
   Where \(x_i\) is the value of the property at time \(i\) and \(n\) is the total number of data points.

2. **Standard Deviation**:  
   The standard deviation measures the dispersion of the values from the mean. A low value indicates relatively constant data, while a high value suggests significant fluctuations. It is calculated as:  
   $$ \text{Standard Deviation} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \text{Mean})^2} $$

3. **Range**:  
   The range is the difference between the maximum and minimum values observed during the simulation. It is calculated with the formula:  
   $$ \text{Range} = \text{Maximum Value} - \text{Minimum Value} $$

## Analysis of Properties

Below are examples of how these values were calculated for **RMSD** (Root Mean Square Deviation), and the same approach applies to other properties.

### 1. RMSD (Root Mean Square Deviation)
RMSD measures the change in the structure of the molecule over time, compared to its initial structure. A low value indicates that the structure is stable, while a high value suggests significant changes in the structure. In our analysis, the following values were calculated for the `rmsd_25.xvg` file:

- **RMSD Mean**: 0.865 √Ö
- **RMSD Standard Deviation**: 0.349 √Ö
- **RMSD Range**: 1.925 √Ö

These values suggest that the simulated system (file 25.xvg) maintains a relatively stable structure, although with some fluctuations. Comparing this to the `rmsd_aldoseREF.xvg` file reveals similar values, suggesting that both simulations exhibit similar structural behavior, with slight differences. This information is crucial for evaluating the structural stability of the molecules during the simulation.

### 2. RMSF (Root Mean Square Fluctuation)
RMSF provides information about the mobility of each atom within the molecule during the simulation. This calculation is useful for identifying regions that might be more flexible or stable in the structure, which is relevant for drug design or the study of dynamic proteins. RMSF values are calculated similarly to RMSD, allowing us to identify which regions of the molecule are more mobile.

### 3. Radius of Gyration (Rg)
The radius of gyration is a measure of the compactness of a molecule. A low \( R_g \) value indicates a more compact structure, while a higher value suggests a more extended structure. This analysis is crucial for understanding the conformation of proteins or macromolecular complexes.

### 4. HBond (Hydrogen Bonds)
Hydrogen bonds are key interactions that stabilize the three-dimensional structures of proteins and other biomolecules. The analysis of hydrogen bonds during the simulation provides information about the stability of the molecular structure. Significant fluctuations in HBond values may indicate conformational changes or the breaking of key interactions.

### 5. SASA (Solvent Accessible Surface Area)
SASA measures the portion of the molecule exposed to the solvent, which is important for solubility, molecular-solvent interactions, and understanding protein activity. This analysis is especially relevant in the study of proteins and other biomolecules, as the exposed surface areas may influence biological activity.

### 6. Comparison of Kinetic and Bond Energy
Comparing the kinetic energies of different simulations provides insight into how the system's dynamics vary over time. Significant fluctuations in kinetic energy may indicate phase transitions or conformational changes. On the other hand, bond energy provides information about the stability of the system. Lower bond energy generally indicates a more stable system, while higher bond energy suggests unfavorable interactions.

## Application of the Results

The statistical calculations, such as the mean, standard deviation, and range, are essential tools for interpreting the simulation results. These metrics not only provide a quantitative view of the variability within the system but are also crucial for optimizing simulation conditions or identifying unusual behaviors that may be of interest in drug design, structural biology, or computational chemistry.

The analysis of these structural and energetic properties, combined with statistical calculations, enables a deeper understanding of molecular dynamics, helping to detect patterns, anomalies, or trends that enhance our ability to predict the behavior of molecules under specific conditions. This information is vital for fields such as molecular design, protein engineering, and the study of biomolecular complexes, where stability and interactions play key roles in biological and chemical processes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to read the .xvg files and extract RMSD values
def read_xvg(file_name):
    data = np.loadtxt(file_name, comments=["@", "#"])  # Ignores comment lines
    return data[:, 1]  # Assumes RMSD is in the second column

# Read the RMSD files
rmsd_25 = read_xvg('rmsd_25.xvg')
rmsd_aldoseREF = read_xvg('rmsd_aldoseREF.xvg')

# Calculate the metrics
def calculate_metrics(rmsd_values):
    rmsd_avg = np.mean(rmsd_values)
    rmsd_std = np.std(rmsd_values)
    rmsd_range = np.max(rmsd_values) - np.min(rmsd_values)
    return rmsd_avg, rmsd_std, rmsd_range

# Metrics for each file
metrics_25 = calculate_metrics(rmsd_25)
metrics_aldoseREF = calculate_metrics(rmsd_aldoseREF)

# Print the results
print(f"RMSD Average (rmsd_25): {metrics_25[0]:.3f} √Ö")
print(f"Standard Deviation (rmsd_25): {metrics_25[1]:.3f} √Ö")
print(f"RMSD Range (rmsd_25): {metrics_25[2]:.3f} √Ö")

print(f"RMSD Average (rmsd_aldoseREF): {metrics_aldoseREF[0]:.3f} √Ö")
print(f"Standard Deviation (rmsd_aldoseREF): {metrics_aldoseREF[1]:.3f} √Ö")
print(f"RMSD Range (rmsd_aldoseREF): {metrics_aldoseREF[2]:.3f} √Ö")

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(rmsd_25, label=f'RMSD 25 (Avg: {metrics_25[0]:.2f} √Ö)', color='blue')
plt.plot(rmsd_aldoseREF, label=f'RMSD aldoseREF (Avg: {metrics_aldoseREF[0]:.2f} √Ö)', color='red')
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (√Ö)')
plt.legend()
plt.title('RMSD 3G5E')
plt.show()
```

**Comparison of Kinetic Energy 3G5E**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Upload both .xvg files if not already in Google Colab
from google.colab import files
uploaded = files.upload()  # Select energy_output_kinetic_25.xvg and energy_output_kinetic_aldoseREF.xvg

# Function to read and extract time and energy data from an .xvg file
def read_xvg(file_name):
    time = []
    kinetic_energy = []
    total_energy = []
    with open(file_name, "r") as file:
        for line in file:
            if line.startswith("#") or line.startswith("@"):
                continue  # Skip comments and metadata
            data = line.split()
            time.append(float(data[0]))
            kinetic_energy.append(float(data[1]))
            total_energy.append(float(data[2]))  # Assuming total energy is in the third column

    # Convert to numpy arrays
    time = np.array(time)
    kinetic_energy = np.array(kinetic_energy)
    total_energy = np.array(total_energy)

    return time, kinetic_energy, total_energy

# Function to calculate statistics
def calculate_statistics(energy_data):
    average = np.mean(energy_data)
    std_error = np.std(energy_data) / np.sqrt(len(energy_data))
    rmsd = np.sqrt(np.mean((energy_data - average)**2))
    tot_drift = energy_data[-1] - energy_data[0]

    return average, std_error, rmsd, tot_drift

# Read data from both files
time_25, kinetic_25, total_25 = read_xvg("energy_output_kinetic_25.xvg")
time_aldose, kinetic_aldose, total_aldose = read_xvg("energy_output_kinetic_aldoseREF.xvg")

# Calculate statistics for kinetic and total energy
kinetic_avg_25, kinetic_err_25, kinetic_rmsd_25, kinetic_tot_drift_25 = calculate_statistics(kinetic_25)
total_avg_25, total_err_25, total_rmsd_25, total_tot_drift_25 = calculate_statistics(total_25)

kinetic_avg_aldose, kinetic_err_aldose, kinetic_rmsd_aldose, kinetic_tot_drift_aldose = calculate_statistics(kinetic_aldose)
total_avg_aldose, total_err_aldose, total_rmsd_aldose, total_tot_drift_aldose = calculate_statistics(total_aldose)

# Print the results for both files
print("Kinetic Energy (25.xvg):")
print(f"Average: {kinetic_avg_25:.2f} kJ/mol")
print(f"Err.Est.: {kinetic_err_25:.2f} kJ/mol")
print(f"RMSD: {kinetic_rmsd_25:.2f} kJ/mol")
print(f"Tot-Drift: {kinetic_tot_drift_25:.2f} kJ/mol")

print("\nTotal Energy (25.xvg):")
print(f"Average: {total_avg_25:.2f} kJ/mol")
print(f"Err.Est.: {total_err_25:.2f} kJ/mol")
print(f"RMSD: {total_rmsd_25:.2f} kJ/mol")
print(f"Tot-Drift: {total_tot_drift_25:.2f} kJ/mol")

print("\nKinetic Energy (aldoseREF.xvg):")
print(f"Average: {kinetic_avg_aldose:.2f} kJ/mol")
print(f"Err.Est.: {kinetic_err_aldose:.2f} kJ/mol")
print(f"RMSD: {kinetic_rmsd_aldose:.2f} kJ/mol")
print(f"Tot-Drift: {kinetic_tot_drift_aldose:.2f} kJ/mol")

print("\nTotal Energy (aldoseREF.xvg):")
print(f"Average: {total_avg_aldose:.2f} kJ/mol")
print(f"Err.Est.: {total_err_aldose:.2f} kJ/mol")
print(f"RMSD: {total_rmsd_aldose:.2f} kJ/mol")
print(f"Tot-Drift: {total_tot_drift_aldose:.2f} kJ/mol")

# Plotting the kinetic energies for comparison
plt.figure(figsize=(10, 6))
plt.plot(time_25, kinetic_25, label='Kinetic Energy (25.xvg)', color='blue')
plt.plot(time_aldose, kinetic_aldose, label='Kinetic Energy (aldoseREF.xvg)', color='red')
plt.xlabel('Time (ns)')
plt.ylabel('Kinetic Energy (kJ/mol)')
plt.title('Comparison of Kinetic Energy 3G5E')
plt.legend()
plt.grid(False)  # Turn off grid lines
plt.gca().set_facecolor("white")  # Set plot background to white
plt.xlim(0, 100)  # Limit the X-axis to 100 ns
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Upload the .xvg files
from google.colab import files
uploaded = files.upload()  # Select energy_output_bond_25.xvg and energy_output_bond_aldoseREF.xvg

# Function to read and extract data from .xvg files
def read_xvg(file_name):
    time = []
    bond_energy = []
    u_b_energy = []
    proper_dih_energy = []
    improper_dih_energy = []
    lj14_energy = []
    coulomb14_energy = []
    coulomb_sr_energy = []
    coulomb_recip_energy = []
    total_energy = []

    with open(file_name, "r") as file:
        for line in file:
            if line.startswith("#") or line.startswith("@"):
                continue  # Skip comments and metadata
            data = line.split()
            time.append(float(data[0]))
            bond_energy.append(float(data[1]))
            u_b_energy.append(float(data[2]))
            proper_dih_energy.append(float(data[3]))
            improper_dih_energy.append(float(data[4]))
            lj14_energy.append(float(data[5]))
            coulomb14_energy.append(float(data[6]))
            coulomb_sr_energy.append(float(data[7]))
            coulomb_recip_energy.append(float(data[8]))
            total_energy.append(float(data[9]))

    # Convert to numpy arrays
    time = np.array(time)
    bond_energy = np.array(bond_energy)
    u_b_energy = np.array(u_b_energy)
    proper_dih_energy = np.array(proper_dih_energy)
    improper_dih_energy = np.array(improper_dih_energy)
    lj14_energy = np.array(lj14_energy)
    coulomb14_energy = np.array(coulomb14_energy)
    coulomb_sr_energy = np.array(coulomb_sr_energy)
    coulomb_recip_energy = np.array(coulomb_recip_energy)
    total_energy = np.array(total_energy)

    return time, bond_energy, u_b_energy, proper_dih_energy, improper_dih_energy, lj14_energy, coulomb14_energy, coulomb_sr_energy, coulomb_recip_energy, total_energy

# Function to calculate statistics (Average, Err.Est., RMSD, Tot-Drift)
def calculate_statistics(energy_data):
    average = np.mean(energy_data)
    std_error = np.std(energy_data) / np.sqrt(len(energy_data))
    rmsd = np.sqrt(np.mean((energy_data - average)**2))
    tot_drift = energy_data[-1] - energy_data[0]

    return average, std_error, rmsd, tot_drift

# Read data from both files
time_25, bond_25, u_b_25, proper_dih_25, improper_dih_25, lj14_25, coulomb14_25, coulomb_sr_25, coulomb_recip_25, total_25 = read_xvg("energy_output_bond_25.xvg")
time_aldose, bond_aldose, u_b_aldose, proper_dih_aldose, improper_dih_aldose, lj14_aldose, coulomb14_aldose, coulomb_sr_aldose, coulomb_recip_aldose, total_aldose = read_xvg("energy_output_bond_aldoseREF.xvg")

# Calculate statistics for each energy term for both files
def calculate_and_plot_statistics(time_25, bond_25, u_b_25, proper_dih_25, improper_dih_25, lj14_25, coulomb14_25, coulomb_sr_25, coulomb_recip_25, total_25, time_aldose, bond_aldose, u_b_aldose, proper_dih_aldose, improper_dih_aldose, lj14_aldose, coulomb14_aldose, coulomb_sr_aldose, coulomb_recip_aldose, total_aldose):
    # Calculate statistics for each energy term for the 25 file
    bond_avg_25, bond_err_25, bond_rmsd_25, bond_tot_drift_25 = calculate_statistics(bond_25)
    u_b_avg_25, u_b_err_25, u_b_rmsd_25, u_b_tot_drift_25 = calculate_statistics(u_b_25)
    proper_dih_avg_25, proper_dih_err_25, proper_dih_rmsd_25, proper_dih_tot_drift_25 = calculate_statistics(proper_dih_25)
    improper_dih_avg_25, improper_dih_err_25, improper_dih_rmsd_25, improper_dih_tot_drift_25 = calculate_statistics(improper_dih_25)
    lj14_avg_25, lj14_err_25, lj14_rmsd_25, lj14_tot_drift_25 = calculate_statistics(lj14_25)
    coulomb14_avg_25, coulomb14_err_25, coulomb14_rmsd_25, coulomb14_tot_drift_25 = calculate_statistics(coulomb14_25)
    coulomb_sr_avg_25, coulomb_sr_err_25, coulomb_sr_rmsd_25, coulomb_sr_tot_drift_25 = calculate_statistics(coulomb_sr_25)
    coulomb_recip_avg_25, coulomb_recip_err_25, coulomb_recip_rmsd_25, coulomb_recip_tot_drift_25 = calculate_statistics(coulomb_recip_25)
    total_avg_25, total_err_25, total_rmsd_25, total_tot_drift_25 = calculate_statistics(total_25)

    # Calculate statistics for each energy term for the aldoseREF file
    bond_avg_aldose, bond_err_aldose, bond_rmsd_aldose, bond_tot_drift_aldose = calculate_statistics(bond_aldose)
    u_b_avg_aldose, u_b_err_aldose, u_b_rmsd_aldose, u_b_tot_drift_aldose = calculate_statistics(u_b_aldose)
    proper_dih_avg_aldose, proper_dih_err_aldose, proper_dih_rmsd_aldose, proper_dih_tot_drift_aldose = calculate_statistics(proper_dih_aldose)
    improper_dih_avg_aldose, improper_dih_err_aldose, improper_dih_rmsd_aldose, improper_dih_tot_drift_aldose = calculate_statistics(improper_dih_aldose)
    lj14_avg_aldose, lj14_err_aldose, lj14_rmsd_aldose, lj14_tot_drift_aldose = calculate_statistics(lj14_aldose)
    coulomb14_avg_aldose, coulomb14_err_aldose, coulomb14_rmsd_aldose, coulomb14_tot_drift_aldose = calculate_statistics(coulomb14_aldose)
    coulomb_sr_avg_aldose, coulomb_sr_err_aldose, coulomb_sr_rmsd_aldose, coulomb_sr_tot_drift_aldose = calculate_statistics(coulomb_sr_aldose)
    coulomb_recip_avg_aldose, coulomb_recip_err_aldose, coulomb_recip_rmsd_aldose, coulomb_recip_tot_drift_aldose = calculate_statistics(coulomb_recip_aldose)
    total_avg_aldose, total_err_aldose, total_rmsd_aldose, total_tot_drift_aldose = calculate_statistics(total_aldose)

    # Generate comparative plots
    plt.figure(figsize=(10, 6))
    plt.plot(time_25, bond_25, label="Bond Energy 25.xvg", color="blue")
    plt.plot(time_aldose, bond_aldose, label="Bond Energy aldoseREF.xvg", color="red")
    plt.xlabel("Time (ns)")
    plt.ylabel("Bond Energy (kJ/mol)")
    plt.title("Bond Energy 3G5E")
    plt.legend()
    plt.show()

    # Print results for each energy term
    def print_results(label, avg, err, rmsd, tot_drift):
        print(f"{label}:")
        print(f"Average: {avg:.2f} kJ/mol")
        print(f"Err.Est.: {err:.2f} kJ/mol")
        print(f"RMSD: {rmsd:.2f} kJ/mol")
        print(f"Tot-Drift: {tot_drift:.2f} kJ/mol\n")

    print_results("Bond Energy (25.xvg)", bond_avg_25, bond_err_25, bond_rmsd_25, bond_tot_drift_25)
    print_results("U-B Energy (25.xvg)", u_b_avg_25, u_b_err_25, u_b_rmsd_25, u_b_tot_drift_25)
    print_results("Proper Dih. (25.xvg)", proper_dih_avg_25, proper_dih_err_25, proper_dih_rmsd_25, proper_dih_tot_drift_25)
    print_results("Improper Dih. (25.xvg)", improper_dih_avg_25, improper_dih_err_25, improper_dih_rmsd_25, improper_dih_tot_drift_25)
    print_results("LJ-14 (25.xvg)", lj14_avg_25, lj14_err_25, lj14_rmsd_25, lj14_tot_drift_25)
    print_results("Coulomb-14 (25.xvg)", coulomb14_avg_25, coulomb14_err_25, coulomb14_rmsd_25, coulomb14_tot_drift_25)
    print_results("Coulomb SR (25.xvg)", coulomb_sr_avg_25, coulomb_sr_err_25, coulomb_sr_rmsd_25, coulomb_sr_tot_drift_25)
    print_results("Coulomb Recip. (25.xvg)", coulomb_recip_avg_25, coulomb_recip_err_25, coulomb_recip_rmsd_25, coulomb_recip_tot_drift_25)
    print_results("Total Energy (25.xvg)", total_avg_25, total_err_25, total_rmsd_25, total_tot_drift_25)

    print_results("Bond Energy (aldoseREF.xvg)", bond_avg_aldose, bond_err_aldose, bond_rmsd_aldose, bond_tot_drift_aldose)
    print_results("U-B Energy (aldoseREF.xvg)", u_b_avg_aldose, u_b_err_aldose, u_b_rmsd_aldose, u_b_tot_drift_aldose)
    print_results("Proper Dih. (aldoseREF.xvg)", proper_dih_avg_aldose, proper_dih_err_aldose, proper_dih_rmsd_aldose, proper_dih_tot_drift_aldose)
    print_results("Improper Dih. (aldoseREF.xvg)", improper_dih_avg_aldose, improper_dih_err_aldose, improper_dih_rmsd_aldose, improper_dih_tot_drift_aldose)
    print_results("LJ-14 (aldoseREF.xvg)", lj14_avg_aldose, lj14_err_aldose, lj14_rmsd_aldose, lj14_tot_drift_aldose)
    print_results("Coulomb-14 (aldoseREF.xvg)", coulomb14_avg_aldose, coulomb14_err_aldose, coulomb14_rmsd_aldose, coulomb14_tot_drift_aldose)
    print_results("Coulomb SR (aldoseREF.xvg)", coulomb_sr_avg_aldose, coulomb_sr_err_aldose, coulomb_sr_rmsd_aldose, coulomb_sr_tot_drift_aldose)
    print_results("Coulomb Recip. (aldoseREF.xvg)", coulomb_recip_avg_aldose, coulomb_recip_err_aldose, coulomb_recip_rmsd_aldose, coulomb_recip_tot_drift_aldose)
    print_results("Total Energy (aldoseREF.xvg)", total_avg_aldose, total_err_aldose, total_rmsd_aldose, total_tot_drift_aldose)

# Call the function to calculate and plot
calculate_and_plot_statistics(time_25, bond_25, u_b_25, proper_dih_25, improper_dih_25, lj14_25, coulomb14_25, coulomb_sr_25, coulomb_recip_25, total_25, time_aldose, bond_aldose, u_b_aldose, proper_dih_aldose, improper_dih_aldose, lj14_aldose, coulomb14_aldose, coulomb_sr_aldose, coulomb_recip_aldose, total_aldose)
```
